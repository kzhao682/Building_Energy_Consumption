```python
#function to query building egauge data and pickle to file
def get_egauge_hours(dataid, filename=None):
    table = pd.read_sql_query("""SELECT * FROM university.electricity_egauge_hours as a
                WHERE dataid={} AND localhour BETWEEN '01-01-2014' AND '05-01-2017'""".format(dataid), conn)
    table = table.sort_values('localhour', ascending=True)
    
    if filename:
        pickle.dump(table, open('pickle_files/'+filename,'wb'))
        
    return table
```
